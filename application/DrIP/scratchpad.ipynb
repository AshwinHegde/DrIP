```python
import pandas as pd
```

```python
smiles_df = pd.read_csv('smiles_dictionary.csv', header = None, names = ['drug_name', 'smiles'])
smiles_df.head()
```

```python
import sqlite3
from sqlite3 import Error

def create_connection(db_file):
    """ create a database connection to the SQLite database
        specified by db_file
    
    :param db_file: database file
    :return: Connection object or None
    """
    conn = None
    try:
        conn = sqlite3.connect(db_file)
        return conn
    except Error as e:
        print(e)
 
    return conn

def create_table(conn, create_table_sql):
    """ create a table from the create_table_sql statement

    :param conn: Connection object
    :param create_table_sql: a CREATE TABLE statement
    :return:
    """
    try:
        c = conn.cursor()
        c.execute(create_table_sql)
    except Error as e:
        print(e)
```

```python
db_file = 'smiles_db.db'

sql_create_table = """ CREATE TABLE IF NOT EXISTS drug_smiles (
                                        drug_name text PRIMARY KEY,
                                        smiles text NOT NULL
                                    ); """

conn = create_connection(db_file)
if conn is not None:
    create_table(conn, sql_create_table)
else:
    print('Cannot create database.')
```

```python
smiles_df.to_sql('drug_smiles', con = conn, index = False)
```
